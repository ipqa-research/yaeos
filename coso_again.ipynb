```python
import thermo

import numpy as np
```

```python
models = [
    thermo.PRMIX,
    thermo.PR78MIX,
    thermo.SRKMIX
]

tc = [469.7, 507.6]
pc = [33.7 * 1e5, 30.250000000000004 * 1e5]
w = [0.251506, 0.301261]

n = np.array([4.0, 6.0])
nt = np.sum(n)

z = (n / nt).tolist()

v = 1.0
t = 298.15

for idx, model in enumerate(models):
    models[idx] = model(Tcs=tc, Pcs=pc, omegas=w, zs=z, V=v, T=t)
```

```python
coso = models[0]
```

```python
a_exp = []

for model in models:
   a_exp.append(model.A_dep_g / 100)


print("a_exp=", a_exp)
```
